```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

full = cv2.imread('Resources/dog.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)

face = cv2.imread('Resources/face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)

# Comparison methods
methods = [cv2.TM_CCOEFF, cv2.TM_CCOEFF_NORMED, cv2.TM_CCORR, cv2.TM_CCORR_NORMED, cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]

# Perform comparison on image
for m in methods:
    # Create a copy
    full_copy = full.copy()
    method = m  # Don't need eval() here
    
    # Template Matching
    res = cv2.matchTemplate(full_copy, face, method)
    
    # Find the best match and its location
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    # For methods TM_SQDIFF and TM_SQDIFF_NORMED, use min_loc. Otherwise, use max_loc.
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    
    # Get the dimensions of the template
    h, w = face.shape[:2]
    
    # Define the bottom right corner of the rectangle
    bottom_right = (top_left[0] + w, top_left[1] + h)
    
    # Draw a rectangle around the match
    cv2.rectangle(full_copy, top_left, bottom_right, (255, 0, 0), 10)
    
    # Display the result
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Matching Result')
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('Detected Area')
    
    plt.show()
```
