```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

nadia = cv2.imread('Resources/Nadia_Murad.jpg', 0)
plt.imshow(nadia, cmap='gray')

face_cascade = cv2.CascadeClassifier('Resources/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('Resources/haarcascades/haarcascade_eye.xml')

def detect_face_eyes(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
    
    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 255, 255), 10)
        roi_face = face_img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_face)
        
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(face_img[y:y+h, x:x+w], (ex, ey), (ex+ew, ey+eh), (0, 0, 255), 5)
        
    return face_img

result = detect_face_eyes(nadia)
plt.imshow(result, cmap='gray')
```

```python
# face detection
import cv2

face_cascade = cv2.CascadeClassifier('Resources/haarcascades/haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture('Resources/SETUP.mp4')

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert frame to grayscale
    
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)  # Draw a blue rectangle around each detected face
    
    cv2.imshow('Video Face Detect', frame)
    
    k = cv2.waitKey(1)
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Load the cascade classifier for the object you want to detect
object_cascade = cv2.CascadeClassifier('Resources/haarcascades/haarcascade_frontalface_alt.xml')

cap = cv2.VideoCapture('Resources/SETUP.mp4')

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert frame to grayscale
    
    # Detect objects using the cascade classifier
    objects = object_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in objects:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)  # Draw a green rectangle around each detected object
    
    cv2.imshow('Video Object Detect', frame)
    
    k = cv2.waitKey(1)
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

face_cascade = cv2.CascadeClassifier('Resources/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('Resources/haarcascades/haarcascade_eye.xml')

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert frame to grayscale
    
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)  # Draw a blue rectangle around each detected face
        
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        
        eyes = eye_cascade.detectMultiScale(roi_gray)
        
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)  # Draw a green rectangle around each detected eye

    cv2.imshow('Face and Eye Detection', frame)
    
    key = cv2.waitKey(1)
    if key == ord('q'):  # Press 'q' to exit
        break

cap.release()
cv2.destroyAllWindows()
```
