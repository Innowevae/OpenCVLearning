```python
# Bleanding of img of differnet Size

# Reading First Image and apply color-conveter to RGB From BGR
img1=cv2.imread('puppy.jpg')
img1 =cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

# Reading Second Image and apply color-conveter to RGB From BGR
img2=cv2.imread('copy.jpg')
img2 =cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)


#Resize img
img1=cv2.resize(img1,(1500,1500))
img2=cv2.resize(img2,(600,600))

# chatch the x and y offset o large img 
x_offset = 1500-600
y_offset = 1500-600

rows,cols,channels=img2.shape

# cut the offset size image of img1 
roi = img1[y_offset:1500,y_offset:1500]
plt.imshow(roi)

# Convert img2 to gray and then inverse the output Image
img2gray=cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
plt.imshow(img2gray ,cmap='gray')

#inverse the current img
mask_inv= cv2.bitwise_not(img2gray)
plt.imshow(mask_inv ,cmap='gray')
mask_inv.shape

fg=cv2.bitwise_or(img2,img2,mask=mask_inv)
plt.imshow(fg)

# Final Image
final=cv2.bitwise_or(roi,fg)
plt.imshow(final)
```
