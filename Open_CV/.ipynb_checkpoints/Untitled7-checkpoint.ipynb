```python
import cv2
import time
cap = cv2.VideoCapture('Resources/SETUP.mp4')

if cap.isOpened() == False:
    print('There is some Error')
    
while cap.isOpened():
    ret,frame=cap.read()
    
    if ret == True:
        time.sleep(1/30)
        cv2.imshow('frame',frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        else:
            break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time

# Load the video
cap = cv2.VideoCapture('Resources/SETUP.mp4')

if not cap.isOpened():
    print('There is some Error')

while cap.isOpened():
    ret, frame = cap.read()

    if ret:
        # You can specify the coordinates of the rectangle and its color
        # Here's an example rectangle drawn on the frame:
        # Rectangle coordinates: (x, y, width, height)
        x, y, width, height = 100, 100, 200, 200  # Example coordinates

        # Color in BGR format (Blue, Green, Red)
        color = (0, 255, 0)  # Green

        # Thickness of the rectangle border
        thickness = 2

        # Draw the rectangle on the frame
        cv2.rectangle(frame, (x, y), (x + width, y + height), color, thickness)

        cv2.imshow('frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()

```
